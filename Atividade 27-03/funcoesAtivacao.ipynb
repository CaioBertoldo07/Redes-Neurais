### Função de ativação sigmoide

$$f(x) = \frac{1}{1 + e^{-\beta \cdot x}}$$

- Implemente uma função em Python que retorne o valor da função sigmóide para uma dada entrada
- Nesta função, considere beta como parâmetro nomeado com valor igual a 1.0
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função sigmoide
- Em um mesmo gráfico, ilustre 3 valores distintos para $\beta$:
  - $\beta = 0.7$ na cor laranja
  - $\beta = 0.5$ na cor verde
  - $\beta = 0.3$ na cor vermelha
  - Use legendas no gráfico

```python
def sigmoid(x, beta=1.0):
    return 1 / (1 + np.exp(-beta * x))

def sigmoidDerivative(x, beta=1.0):
    s = sigmoid(x, beta)
    return beta * s * (1 - s)

betaValues = [0.7, 0.5, 0.3]
colors = ['orange','green', 'red']

plt.figure(figsize=(8, 5))
for beta, color in zip(betaValues, colors):
    y = sigmoid(x, beta)
    plt.plot(x, y, label=f'β = {beta}', color=color)

plt.xlabel("x")
plt.ylabel("Sigmoide(x)")
plt.title("Função Sigmoide para diferentes valores de β")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
for beta, color in zip(betaValues, colors):
    y = sigmoidDerivative(x, beta)
    plt.plot(x, y, label=f'β = {beta}', color=color)

plt.xlabel("x")
plt.ylabel("Sigmoide(x)")
plt.title("Função Derivada da Sigmoide para diferentes valores de β")
plt.legend()
plt.grid()
plt.show()
```

### Função de Ativação ReLU (Unidade Linear Retificada)

$$f(x) = max(0,x)$$

- Implemente uma função em Python que retorne o valor da função ReLU para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função ReLU

```python
def relu(x):
    return np.maximum(0, x)

def reluDerivative(x):
    return np.where(x > 0, 1, 0)

plt.figure(figsize=(8, 5))
plt.plot(x, relu(x), label='ReLU', color='blue')

plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.title("Função de Ativação ReLU")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, reluDerivative(x), label='ReLU', color='blue')

plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.title("Função de Ativação Derivada ReLU")
plt.legend()
plt.grid()
plt.show()
```

### Função de ativação linear (purelin)

$$f(x) = x$$

### Função de ativação Tangente Hiperbólica

$$f(x) = tanh(x)$$

- Implemente uma função em Python que retorne o valor da função tangente hiperbólica para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função em questão
- Utilize bibliotecas para auxiliar nesta tarefa
- Considere o intervalo (-1,1)

```python
def tanH(x):
    return np.tanh(x)

def tanHDerivative(x):
    return 1 - np.tanh(x)**2

plt.figure(figsize=(8, 5))
plt.plot(x, tanH(x), label='tanh(x)', color='blue')

plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.title("Função de Ativação Tangente Hiperbólica")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, tanHDerivative(x), label='tanh(x)', color='blue')

plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.title("Função de Ativação Derivada da Tangente Hiperbólica")
plt.legend()
plt.grid()
plt.show()
```

### Função de Ativação Gaussiana

$$f(x) = \exp \left( \frac{-(x - \mu)^2}{2\sigma^2} \right)$$

- $\mu$ é o ponto médio (centro) da curva
- $\sigma$ denota o raio de abertura
- Implemente uma função em Python que retorne o valor da função gaussiana para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função em questão


Considere também os seguintes valores ao plotar a função

- Parâmetros nomeados $\mu = 0$ e $\sigma = 1$
- $\mu = 0, \sigma = 0.3$
- $\mu = 0, \sigma = 0.5$
- $\mu = 0, \sigma = 0.7$
- $\mu = 0, \sigma = 0.9$
- $\mu = 0, \sigma = 1.1$

```python
def gaussian(x, mu=0, sigma=1):
    return np.exp(-((x - mu) ** 2) / (2 * sigma ** 2))

def gaussianDerivative(x, mu=0, sigma=1):
    return -((x - mu) / (sigma ** 2)) * gaussian(x, mu, sigma)

sigmaValues = [0.3, 0.5, 0.7, 0.9, 1.1]
colors = ['red', 'green', 'blue', 'orange', 'purple']

plt.figure(figsize=(8, 5))
for sigma, color in zip(sigmaValues, colors):
    y = gaussian(x, mu=0, sigma=sigma)
    plt.plot(x, y, label=f'σ = {sigma}', color=color)
    
plt.xlabel("x")
plt.ylabel("Gaussiana(x)")
plt.title("Função de Ativação Gaussiana para diferentes valores de σ")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
for sigma, color in zip(sigmaValues, colors):
    y = gaussianDerivative(x, mu=0, sigma=sigma)
    plt.plot(x, y, label=f'σ = {sigma}', color=color)
    
plt.xlabel("x")
plt.ylabel("Gaussiana(x)")
plt.title("Função de Ativação Derivada da Gaussiana para diferentes valores de σ")
plt.legend()
plt.grid()
plt.show()
```

### Função de Ativação Leaky ReLU

$$f(x) = \left\{\begin{array}{cc} 0.01\cdot x & x < 0\\
x & x \geq 0
\end{array}
\right.$$

- Implemente uma função em Python que retorne o valor da função Leaky ReLU para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função Leaky ReLU (consulte a literatura!)

```python
def leakyRelu(x, alpha=0.01):
    return np.where(x >= 0, x, alpha * x)

def leakyReluDerivative(x, alpha=0.01):
    return np.where(x >= 0, 1, alpha)

plt.figure(figsize=(8, 5))
plt.plot(x, leakyRelu(x), label='Leaky ReLU', color='blue')

plt.xlabel("x")
plt.ylabel("Leaky ReLU(x)")
plt.title("Função de Ativação Leaky ReLU")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, leakyReluDerivative(x), label='Leaky ReLU', color='blue')

plt.xlabel("x")
plt.ylabel("Leaky ReLU(x)")
plt.title("Função de Ativação Derivada Leaky ReLU")
plt.legend()
plt.grid()
plt.show()
```
